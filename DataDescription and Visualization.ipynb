```python
import os
import pandas as pd

#Load the filtered dataset up to 2017

# Paths
raw_folder = r"C:\Users\zigar\Desktop\library-checkouts-project\data\raw"
cleaned_folder = r"C:\Users\zigar\Desktop\library-checkouts-project\data\cleaned"
input_csv = os.path.join(raw_folder, "checkouts_by_title.csv")
output_csv = os.path.join(cleaned_folder, "checkouts_up_to_2017.csv")

# Remove previous output if exists
if os.path.exists(output_csv):
    os.remove(output_csv)

# Process CSV in chunks
chunksize = 1_000_000  # adjust based on your RAM
total_filtered = 0

for i, chunk in enumerate(pd.read_csv(input_csv, chunksize=chunksize)):
    # Filter rows
    chunk_filtered = chunk[chunk['CheckoutYear'] <= 2017]
    
    if not chunk_filtered.empty:
        # Append to output CSV
        chunk_filtered.to_csv(output_csv, mode='a', index=False,
                              header=not os.path.exists(output_csv))
        total_filtered += len(chunk_filtered)
    
    print(f"Chunk {i+1} processed, total filtered rows so far: {total_filtered}")

print(f"\nFiltered dataset saved at: {output_csv}")
print(f"Total rows up to 2017: {total_filtered}")
```
